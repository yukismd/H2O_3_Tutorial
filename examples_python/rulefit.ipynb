```python
import h2o
print(h2o.__version__)
from h2o.estimators import H2ORuleFitEstimator
```

```python
h2o.init()
```

```python
f = "https://h2oai-jp-public.s3.ap-northeast-1.amazonaws.com/sample_data/UCI_CreditCard/UCI_Credit_Card3.csv"
df = h2o.import_file(path=f, col_types={'ID': "enum", 'default_payment_next_month': "enum"})
df.shape
```

```python
df.describe()
```

```python
# データ分割
frame_train, frame_test = df.split_frame(ratios=[0.8], seed=12345)
frame_train.shape, frame_test.shape
```

```python
# カラムの役割
x_cols = ['SEX','EDUCATION','MARRIAGE','AGE',
     'PAY_1','PAY_2','PAY_3','PAY_4','PAY_5','PAY_6',
     'BILL_AMT1','BILL_AMT2','BILL_AMT3','BILL_AMT4','BILL_AMT5','BILL_AMT6',
     'PAY_AMT1','PAY_AMT2','PAY_AMT3','PAY_AMT4','PAY_AMT5','PAY_AMT6']
y_col = 'default_payment_next_month'
```

### モデル作成

```python
rfit = H2ORuleFitEstimator(seed=123)
rfit.train(x=x_cols, y=y_col, training_frame=frame_train, validation_frame=frame_test)
```

```python
# 学習とテストデータにおけるAUC
rfit.auc(), rfit.auc(valid=True)
```

```python
# ハイパーパラメータ（デフォルト設定）の確認
rfit.actual_params
```

### ルールの確認

```python
# 特徴量重要度（Lassoの回帰係数）とルール（交互作用）の表示
rfit.rule_importance()
```

`M0T14N15, M0T4N15, M0T22N15, M0T30N15`といった形で表示されているのは同じルールが複数の木の間で作成されたため。同じルールなので一つのみLassoに投入される  
（あえて全て投入する場合はハイパーパラメータに`remove_duplicates=False`を指定）

```python
# オブザベーションが各ルールに該当するか。該当すれば1
rule_observation = rfit.predict_rules(frame_test, ['M0T43N20', 'M0T42N16', 'M0T28N21', 'M0T41N18', 'M0T37N21'])   # 上位5ルール
print(rule_observation.shape)
rule_observation
```

```python
# ルール"M0T28N21"に該当しているindex=2のオブザベーションを表示してみる
# ルール"M0T28N21" : (PAY_2 < 1.5) & (PAY_3 >= 1.5) & (PAY_AMT1 >= 853.5 or PAY_AMT1 is NA)
frame_test[2,:]
```

Rule Importanceの'support'とは、各ルールに該当するオブザベーションの割合

```python
rfit.predict_rules(frame_train, ['M0T43N20']).mean()
```

```python
rfit.rule_importance().as_data_frame().iloc[0,:]
```

```python
rule = rfit.rule_importance().as_data_frame()    # pandas dataframeへ変換
rule
```

```python
rule.to_csv('../tmp/rule.csv')
```

### ルール生成の設定を変えてみる

```python
rfit2 = H2ORuleFitEstimator(seed=123,
                            min_rule_length=1,    # 分岐の深さの最小値
                            max_rule_length=2,   # 分岐の深さの最大値（1次の交互作用まで）
                            max_num_rules=50,  # 最大で50ルールまでは作成可能
                           #remove_duplicates=False,
                           )
rfit2.train(x=x_cols, y=y_col, training_frame=frame_train, validation_frame=frame_test)
```

```python
# 学習とテストデータにおけるAUC
rfit2.auc(), rfit2.auc(valid=True)
```

```python
# 特徴量重要度（Lassoの回帰係数）とルール（交互作用）の表示
rfit2_rule = rfit2.rule_importance().as_data_frame()
print(rfit2_rule.shape)
rfit2_rule
```
